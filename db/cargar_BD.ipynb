```python
# pip install pandas sqlalchemy pymysql cryptography

# Cargar librerias 
import pandas as pd
from sqlalchemy import create_engine, text
```

```python
# Datos de Conexion BD
usuario = 'mi_usuario'
contrasena = 'mi_password'
host = 'localhost'
puerto = 3306
base_de_datos = 'mi_base_de_datos'
```

```python
# Cargar data y vista previa
df = pd.read_csv('DataSetFinalProyectoGradoMIAD.csv') 
df.head()
```

```python
# Modificacion a BD
# df.rename(columns={'Porc_part_Propósito general':'Porc_part_Proposito_general'}, inplace=True)
```

```python
# Crear conexion BD
engine = create_engine(f'mysql+pymysql://{usuario}:{contrasena}@{host}:{puerto}/{base_de_datos}')
```

```python
# Revisar conexion
engine.connect().execute(text("SHOW DATABASES")).fetchall()
```

```python
# crear tabla con datos
df.to_sql('dataSetMIAD', con=engine, if_exists='replace', index=False)
```
